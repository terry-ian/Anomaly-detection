```python
#找出全部游戏
time.sleep(2) 
soup = BeautifulSoup(browser.page_source, "html.parser")
game=soup.findAll('a', {"class": "pointer"})
gamealltextgg = list()
for gameall in game :
    if len(gameall.text) != 0 :
        gamealltextgg.append(gameall.text)  
#显示前五款游戏
gamealltextgg[:5]
```

```python
#收集DATA
gamedata=[]
```

```python
#爬蟲主程式   #gamealltextgg[12:]
#i=-1
for gamenumber in gamealltextgg:
    #i=i+1
    #if i<12 : continue
    
    time.sleep(2)
    #2款游戏未上线
    if gamenumber == gamealltextgg[24] or gamenumber == gamealltextgg[25]: continue
    #预防突然跳出截止下注和干扰
    if (gamenumber != gamealltextgg[0]) and (gamenumber != gamealltextgg[8]):
        timenumber=soup.find('div', {"class": "undefined lottery_timeBox left"}).find("b").text[6:8]
        if timenumber == '--': time.sleep(10)
        elif float(timenumber)<10 :time.sleep(20)
    #弹跳示窗
    if len(browser.find_elements_by_css_selector("#close")) != 0 :
        browser.find_element_by_class_name("close").click()
    #点击各游戏进入游戏下注页面
    browser.find_element_by_xpath("//*[contains(text(),'彩票' )]").click( )
    #游戏寻找点击
    browser.find_element_by_xpath("//*[contains(text(),'" + gamenumber + "' )]").click( )
    time.sleep(2) 
    #储存页面
    soup = BeautifulSoup(browser.page_source, "html.parser")
    ###############################################################################
    #大项玩法
    gametypeall = soup.find('div', {"class": "lotteryLeft"}).find('ul')
    gamealltext = list()
    for gameall in gametypeall :
        gamealltext.append(gameall.text)
    #判断长度
    gametypeall_n=len(gamealltext)
    gamealltext
    
    ################################################################################
    for test in gamealltext :
        #预防突然跳出截止下注和干扰
        if gamenumber != gamealltextgg[7]:
            timenumber=soup.find('div', {"class": "undefined lottery_timeBox left"}).find("b").text[6:8]
            if timenumber == '--': time.sleep(10)
            elif float(timenumber)<3 :time.sleep(12)
        #點選五星玩法 '五星', '四星', '后三码', '前三码', '中三码', '二码', '定位胆', '不定胆', '大小单双', '趣味'
        clickgamtype=browser.find_element_by_xpath("//div/ul/li[contains(text(),'" + test + "' )]")
        clickgamtype.click( )
        #储存页面
        soup = BeautifulSoup(browser.page_source, "html.parser")
    
        ###########################################################
        # 小项玩法中各奖金组
        for forplay in range(len(soup.find_all('dl', {"class": "cWay"}))):
            #小项玩法
            gametypeall_small = soup.find_all('dl', {"class": "cWay"})[forplay].find_all("dd")
            gamealltext_samll = list()
            for gameall_small in gametypeall_small :
                gamealltext_samll.append(gameall_small.text)
            
            for j in gamealltext_samll:
                #预防突然跳出截止下注和干扰
                if gamenumber != gamealltextgg[7]:
                    timenumber=soup.find('div', {"class": "undefined lottery_timeBox left"}).find("b").text[6:8]
                    if timenumber == '--': time.sleep(10)
                    elif float(timenumber)<3 :time.sleep(12)
                #点选直选组选的按钮
                clickdir=browser.find_element_by_xpath( "//div/dl/dd[contains(text(),'" + j + "' )]" )
                clickdir.click( )
                #储存页面
                soup = BeautifulSoup(browser.page_source, "html.parser")
                
                #判断有无奖金组
                if len(soup.findAll('option')) == 0 : 
                    continue
                else:
                    #抓取各数值
                    gamename = gamenumber#soup.find('div', {"class": "nav-child-1040"}).find('b')
                    gametype = soup.find('div', {"class": "lotteryLeft"}).find('li', {"class": "hover"})
                    gametype1 = soup.find_all('dl', {"class": "cWay" })[0].find('dt')
                    gametype2= soup.find('dd', {"class": "hover"})
                    pricevalue = soup.find('div', {"class": "fandian"}).find('option') 
                    #soup.find('div', {"class": "fandian"}).find('option', {"value": "2"}).text
                    #显示爬取数据
                    data = [gamename,gametype.text,gametype1.text,gametype2.text,pricevalue.text]
                    gamedata.append(data)
                    print(gamename,'-',gametype.text,'-',gametype1.text,'-',gametype2.text ,'-',pricevalue.text)
        
```

```python
import csv
with open('output.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    title =[['遊戲','遊戲玩法大','選取方式','遊戲玩法小','獎金組']]
    writer.writerows(title)
    # 寫入二維表格
    writer.writerows(gamedata)
```
